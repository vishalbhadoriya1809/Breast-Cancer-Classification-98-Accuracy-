```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
```

```python
data = pd.DataFrame(np.c_[data['data'], data['target']], columns = np.append(data['feature_names'], ['target']))
```

```python
data.to_csv('Breast_Cancer.csv')
```

# Evaluating model

```python
data.shape
```

```python
X = data.iloc[:,0:30]
```

```python
X.head()
```

```python
y = data.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
sns.heatmap(cm, annot = True)
```

```python
print(classification_report(y_test, y_pred))
```

# Improving model

## Step1: Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
standardscaler_X = StandardScaler()
```

```python
X_train = standardscaler_X.fit_transform(X_train)
X_test = standardscaler_X.transform(X_test)
```

```python
X_test
```

## Step:2 Improving HyperParameters

```python
from sklearn.grid_search import GridSearchCV
```

```python
param_grid = [{'C' : [0.1, 0.2, 1 , 2, 10, 100], 'gamma' : [0.1, 1, 0.01, 0.001, 0.0001], 'kernel' : ['rbf']},{'C' : [0.1, 0.2, 1 , 2, 10, 100], 'gamma' : [0.1, 1, 0.01, 0.001, 0.0001], 'kernel' : ['linear']}, {'C' : [0.1, 0.2, 1, 2, 10, 100], 'gamma' : [0.1, 1, 0.01, 0.001, 0.0001], 'kernel' : ['sigmoid']}]
```

```python
grid = GridSearchCV(model, param_grid, refit = True, verbose=4, cv = 10, n_jobs=-1)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_score_
```

```python
grid.error_score
```

```python
grid.best_params_
```

```python
model_imp = SVC(C = 10, gamma=0.01, kernel='rbf', probability=True)
```

```python
model_imp.fit(X_train, y_train)
```

```python
y_pred = model_imp.predict(X_test)
```

```python
cm1 = confusion_matrix(y_test, y_pred)
```

```python
cm1
```

```python
print(classification_report(y_test, y_pred))
```
